## Import Dataset & Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt

# load vgsales data
date_cols = ['release_date', 'last_update']
df = pd.read_csv('data/vgchartz-2024.csv', parse_dates=date_cols)
```

```python
# Deal with duplicates if there is any
df.drop_duplicates()

# Check data shape
print(df.shape)

# Check for missing values
df.info()
```

## Data Cleaning -- Remove All Data With Missing total_sales Values

```python
# drop all rows with missing total_sales values

vgsales_with_totalsales = df[df['total_sales'].notnull()]
vgsales_with_totalsales.info()
```

```python
# Drop all columns that won't be needed for further analysis

indexes = ['title', 'console', 'genre', 'publisher', 'developer', 'total_sales', 'release_date']
vgsales_refined = vgsales_with_totalsales[indexes]
vgsales_refined
```

## Total Game Sales by Year

## Average Game Sales By Month

```python
vg_group_by_month = (vgsales_yrmth
                     .groupby(by=['release_month_name', 'release_month'], as_index=False)
                     .agg(avg_month_sales = ('total_sales', 'mean'))
                     .sort_values(by='release_month')
                    )
vg_group_by_month
```

```python
# Abbreviation for Month name
abbr_month_name = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

bar_color = [ '#00588D' if (month in ['Sep', 'Oct', 'Nov']) else ('#7BBFFC', 0.5) for month in abbr_month_name ]

plt.figure(figsize=(10, 6))
plt.title('Fall is the Sweet Spot for Game Releases!', loc='left', fontsize=16)
plt.ylabel('Sales (in millions)', fontsize=10)
plt.xticks(range(len(abbr_month_name)), abbr_month_name, fontsize=12)
plt.bar(vg_group_by_month['release_month_name'], vg_group_by_month['avg_month_sales'], color=bar_color);
```

```python
bar_color = [ '#A81829' if (month in ['Jul', 'Dec']) else ('#7BBFFC', 0.5) for month in abbr_month_name ]

plt.figure(figsize=(10, 6))
plt.title('Seasonal Slump: Game Sales Struggle in July and December', loc='left', fontsize=16)
plt.ylabel('Sales (in millions)', fontsize=10)
plt.xticks(range(len(abbr_month_name)), abbr_month_name, fontsize=12)
plt.bar(vg_group_by_month['release_month_name'], vg_group_by_month['avg_month_sales'], color=bar_color);
```

## Find the genre of top 100 all times bestseller video games

```python
# group by game title

vg_grouped_by_title = (vgsales_refined
                          .groupby(by='title', as_index=False)
                          .agg(
                              game_total_sales = ('total_sales', 'sum'),
                              game_genre = ('genre', 'max')
                              )
                          .sort_values(by='game_total_sales', ascending=False)
                         )

# get the top 100 all times best seller
best_seller = vg_grouped_by_title.head(n=100)
best_seller
```

```python
# Get the genre of top 100 best seller

popular_genres = best_seller['game_genre'].value_counts()
popular_genres
```

```python
# Show the result in a bar chart

x = popular_genres.index
y = popular_genres.values

bar_color = ['#00588D' if (genre in ['Shooter', 'Sports', 'Action']) else ('#7BBFFC', 0.5) for genre in popular_genres.index]

plt.figure(figsize=(10, 6))
plt.title('The Dominant Game Genres: Shooter, Sports, and Action Lead the Way', fontsize=16)
plt.xlabel('Game Genre', fontsize=12)
plt.ylabel('# of Top 100 Bestseller', fontsize=12)
plt.xticks(rotation=45, fontsize=12)
plt.bar(x, y, align='edge', color=bar_color);
```

## Which console has most video games sold?

```python
# group by console
vg_grouped_by_console = (vgsales_refined
                          .groupby(by='console', as_index=False)
                          .agg(game_total_sales = ('total_sales', 'sum'))
                          .sort_values(by='game_total_sales', ascending=True)
                       )
vg_grouped_by_console['game_total_sales'].median()
```

```python
import matplotlib.patches as mpatches;

# filter consoles who made game_total_sales greater than median(=36.46)
console_to_plot = vg_grouped_by_console[(vg_grouped_by_console['game_total_sales'] >= 36.46)]

# seperate consoles from different companies
ps_consoles = ['PS2','PS3', 'PS', 'PS4', 'PSP', 'PSV']
nintendo_consoles = ['Wii', 'DS', 'GBA', 'GC', '3DS', 'N64', 'SNES', 'NES', 'NS']
xbox_consoles = ['X360', 'XOne', 'XB']


bar_color = ['#00588D' if console in ps_consoles else 
             '#DB444B' if console in nintendo_consoles else 
             '#379A8B' if console in xbox_consoles else 
             ('#B7C6CF', 0.5) for console in console_to_plot['console']]

# Add legend for consoles
legend_ps = mpatches.Patch(color='#00588D', label='PlayStation')
legend_nintendo = mpatches.Patch(color='#DB444B', label='Nintendo')
legend_xbox = mpatches.Patch(color='#379A8B', label='Xbox')

# Add labels to the chart
plt.figure(figsize=(10, 6))
plt.title('PlayStation Occupies 4 Spots Out of The Top 5 Consoles', fontsize=16, loc='left')
plt.xlabel('Sales (in millions)', fontsize=10)
plt.ylabel('Console', fontsize=12)

# Display the legend
plt.legend(handles=[legend_ps, legend_nintendo, legend_xbox], loc='lower right', fontsize=12)

plt.barh(console_to_plot['console'], console_to_plot['game_total_sales'], color=bar_color);
```

## Look for Leading Publishers And Developers

```python
# Find leading developers

vg_grouped_by_dev = (vgsales_refined
                         .groupby(by='developer', as_index=False)
                         .agg(game_total_sales = ('total_sales', 'sum'))
                         .sort_values(by='game_total_sales', ascending=False)
                   )
# filter the top 20 developers

top_developers = vg_grouped_by_dev.head(n=20)
top_developers
```

```python
# Find leading publishers

vg_grouped_by_pub = (vgsales_refined
                         .groupby(by='publisher', as_index=False)
                         .agg(game_total_sales = ('total_sales', 'sum'))
                         .sort_values(by='game_total_sales', ascending=False)
                   )
top_publishers = vg_grouped_by_pub.head(n=20)
top_publishers
```

```python
# Create a DataFrame to highlight companies who are good developers and good publishers at the same time

developers = top_developers['developer'].reset_index(drop=True)
publishers = top_publishers['publisher'].reset_index(drop=True)

combined = pd.DataFrame({'good_developers': developers, 'good_publishers': publishers})

def highlight_overlap(val):
    highlight_colors = {
        'EA': '#006BA2; color: white',
        'Electronic Arts': '#006BA2; color: white',
        'Ubisoft': '#3EBCD2',
        'Konami': '#379A8B; color: white',
        'Capcom': '#EBB434',
        'Namco': '#B4BA39',
        'Rockstar': '#9A607F; color: white',
        'Bethesda': '#D1B07C'
    }

    for key, style in highlight_colors.items():
        if key in val:
            return f'background-color: {style}'
    return ''

combined.style.map(highlight_overlap)
```

## Are Genre, Publisher and Developer Able To Predict Game Sales?

```python
top_genres = popular_genres.head( n = 3 ).index
top_dev = vg_grouped_by_dev['developer'].head( n = 20 )
top_pub = vg_grouped_by_pub['publisher'].head( n = 20 )

vgsales_marked = vgsales_refined.assign( 
                     release_year = vgsales_refined['release_date'].dt.year,
                     is_top_genre = vgsales_refined['genre']
                                              .apply(lambda genre: any( keyword in genre for keyword in top_genres) )
                                              .astype(int),
                     is_top_developer = vgsales_refined['developer']
                                              .apply(lambda dev: any( keyword in dev for keyword in top_dev) if pd.notnull(dev) else False )
                                              .astype(int),
                     is_top_publisher = vgsales_refined['publisher']
                                              .apply(lambda pub: any( keyword in pub for keyword in top_pub) )
                                              .astype(int),
                     competitiveness = lambda df_: df_['is_top_genre'] + df_['is_top_developer'] + df_['is_top_publisher']
                 )

# Noticed some entries have 0 in total_sales, might indicate uncollected data, thus exclude them before further analysis

vgsales_marked = vgsales_marked[vgsales_marked['total_sales'] > 0]
```

```python
# group by competitiveness
vg_grouped_by_comp = (vgsales_marked
                       .groupby(by='competitiveness', as_index=False)
                       .agg( 
                           avg_total_sales = ('total_sales', 'mean'),
                           sum_total_sales = ('total_sales', 'sum'),
                           unique_game_count = ('title', 'nunique')
                       ))
vg_grouped_by_comp
```

```python
bar_colors = ['#5DA4DF', '#3D89C3', '#1270A8', '#00588D']

plt.figure(figsize=(10,8))
plt.title('Higher Competitiveness, Greater Sales', fontsize=16, loc='left')
plt.xlabel('Competitiveness', fontsize=10)
plt.ylabel('Avg Total Sales (in millions)', fontsize=10)
plt.xticks([0, 1, 2, 3])

# Add data labels above bars
bars = plt.bar(vg_grouped_by_comp['competitiveness'], vg_grouped_by_comp['avg_total_sales'], color=bar_colors)
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval - 0.06, f'{yval:.2f}', ha='center', va='bottom', fontsize=12, color='white')
```
